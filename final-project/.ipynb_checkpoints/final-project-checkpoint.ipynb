## Transformação do Dataframe

```python
def transform_data_home(serie):
    data = {'Player': [], 'x': [], 'y': []}

    for i in range(1, 12):
        player = serie['Home_{}_x'.format(i)]
        x = serie['Home_{}_x'.format(i)]
        y = serie['Home_{}_y'.format(i)]

        if pd.notnull(player):
            data['Player'].append('Home_{}'.format(i))
            data['x'].append(x)
            data['y'].append(y)

    ball_x = serie['ball_x']
    ball_y = serie['ball_y']
    if pd.notnull(ball_x) and pd.notnull(ball_y):
        data['Player'].append('Ball')
        data['x'].append(ball_x)
        data['y'].append(ball_y)

    # Cria o novo DataFrame
    df_transformed = pd.DataFrame(data)

    return df_transformed
```

```python
def transform_data_away(serie):
    data = {'Player': [], 'x': [], 'y': []}

    for i in range(15, 26):
        player = serie['Away_{}_x'.format(i)]
        x = serie['Away_{}_x'.format(i)]
        y = serie['Away_{}_y'.format(i)]

        if pd.notnull(player):
            data['Player'].append('Away_{}'.format(i))
            data['x'].append(x)
            data['y'].append(y)

    ball_x = serie['ball_x']
    ball_y = serie['ball_y']
    if pd.notnull(ball_x) and pd.notnull(ball_y):
        data['Player'].append('Ball')
        data['x'].append(ball_x)
        data['y'].append(ball_y)

    # Cria o novo DataFrame
    df_transformed = pd.DataFrame(data)

    return df_transformed
```

## AGM

```python
def gerar_grafo(df_frame):    
    G = nx.Graph()
    
    df_frame = df_frame[df_frame['Player'] != 'Ball']
    
    # Adiciona os nós representando os jogadores
    for _, row in df_frame.iterrows():
        player = row['Player']
        x = row['x']
        y = row['y']
        G.add_node(player, pos=(x, y))

    # Cria as conexões entre as posições e adiciona as distâncias como peso das arestas
    for node1 in G.nodes():
        for node2 in G.nodes():
            if node1 != node2 and not G.has_edge(node1, node2):
                coord1 = G.nodes[node1]['pos']
                coord2 = G.nodes[node2]['pos']
                dist = euclidean(coord1, coord2)
                G.add_edge(node1, node2, weight=dist)

    # Define as posições originais como atributos dos nós
    pos_orig = {player: (data['pos'][0], data['pos'][1]) for player, data in G.nodes(data=True)}
    nx.set_node_attributes(G, pos_orig, 'pos_orig')
    
    return G
```

```python
g_home = gerar_grafo(data_home)
g_away = gerar_grafo(data_away)
```

```python
def plotar_grafo(G):
    
    plt.figure(figsize=(10.6, 6.8))
    plt.xlim(-DIMENSAO_X-5, DIMENSAO_X+5)
    plt.ylim(-DIMENSAO_Y-2, DIMENSAO_Y+2)

    # Plota o grafo utilizando as posições originais
    pos = nx.get_node_attributes(G, 'pos_orig')
    nx.draw(G, pos=pos)
    plt.show()
```

```python
plotar_grafo(g_home)
plotar_grafo(g_away)
```

```python
def gerar_agm(grafo):
    
    # Calcula a AGM usando o algoritmo de Kruskal
    T = nx.minimum_spanning_tree(grafo)
    
    return T
```

```python
def plotar_agm(agm):
    
    pos = nx.get_node_attributes(agm, 'pos_orig')
    fig, ax = mviz.plot_pitch()

    # Plota os nós do grafo
    nx.draw_networkx_nodes(agm, pos=pos, ax=ax)

    # Plota as arestas do grafo
    nx.draw_networkx_edges(agm, pos=pos, ax=ax)

    # Adiciona os rótulos aos jogadores
    labels = {node: node for node in agm.nodes()}
    nx.draw_networkx_labels(agm, pos=pos, labels=labels, font_size=10, font_color='w')

    # Define os limites do gráfico
    ax.set_xlim(-65, 65)
    ax.set_ylim(-35, 35)

    plt.show()
    
```

```python
agm_home = gerar_agm(g_home)
plotar_agm(agm_home)
```

```python
agm_away = gerar_agm(g_away)
plotar_agm(agm_away)
```

```python
def calcular_centralidade(grafo):
    
    # Cálculo das medidas de centralidade
    degree_centrality = nx.degree_centrality(grafo)
    betweenness_centrality = nx.betweenness_centrality(grafo)
    closeness_centrality = nx.closeness_centrality(grafo)
    load_centrality = nx.load_centrality(grafo)

    return degree_centrality, betweenness_centrality, closeness_centrality, load_centrality
```

## Voronoi

```python
def criar_diagrama(df):
    
    df = df[df['Player'] != 'Ball']
    
    # Extrair as coordenadas x e y dos jogadores
    points = df[['x', 'y']].values

   # Calcular o diagrama de Voronoi
    vor = Voronoi(points)

    # Mapeia jogador e area correspondente
    player_region_mapping = {}
    for i, player in enumerate(df['Player']):
        player_region_mapping[player] = vor.point_region[i]

    return vor, player_region_mapping
```

```python
vor,regioes = criar_diagrama(data_home)
```

```python
regioes
```

```python
def plotar_diagrama(vor, df):
    
    if len(df) ==12:
        df = df.drop(index=11)
    
    # Plota o campo
    fig, ax = mviz.plot_pitch()

    # Plota o diagrama de Voronoi
    voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='yellow', line_width=3, point_size=20)

    # Adicionar rótulos aos jogadores
    for i, row in df.iterrows():
        ax.annotate(row['Player'], (row['x'], row['y']))

    # Definir limites do gráfico
    plt.xlim(-65, 65)
    plt.ylim(-35, 35)

    # Mostrar o gráfico
    plt.show()
```

```python
plotar_diagrama(vor, data_home)
```

```python
def gerar_areas(vor, player_region_mapping):
    jogador_area = {}

    for player, region_index in player_region_mapping.items():
        if -1 not in vor.regions[region_index] and len(vor.regions[region_index]) > 0:
            vertices = vor.vertices[vor.regions[region_index]]
            polygon = Polygon(vertices)
            area = polygon.area
        else:
            area = None
        
        jogador_area[player] = area

    return jogador_area
```

## Resultados

#### Centralidade

```python
def gerar_resultados_centralidade(track):
    
    colunas = transform_data_home(track.iloc[0])["Player"][0:11].values
    
    df_grau = pd.DataFrame(columns=colunas)
    df_intermediacao = pd.DataFrame(columns=colunas)
    df_proximidade = pd.DataFrame(columns=colunas)
    df_carga = pd.DataFrame(columns=colunas)
    
    for i in range(0, len(track), 50):
        row = track.iloc[i]
        data = transform_data_home(row)
        g = gerar_grafo(data)
        agm = gerar_agm(g)
        resultados = calcular_centralidade(agm)
        
        df_grau = df_grau.append(resultados[0], ignore_index=True)
        df_intermediacao = df_intermediacao.append(resultados[1], ignore_index=True)
        df_proximidade = df_proximidade.append(resultados[2], ignore_index=True)
        df_carga = df_carga.append(resultados[3], ignore_index=True)
            
    return df_grau, df_intermediacao, df_proximidade, df_carga
```

#### Áreas dos zagueiros

```python
def gerar_resultados_areas_home(track):
    
    colunas = transform_data_home(track.iloc[0])["Player"][0:11].values
    
    df_areas = pd.DataFrame(columns=colunas)
    
    for i in range(0, len(track), 50):
        row = track.iloc[i]
        data = transform_data_home(row)
        vor,regioes = criar_diagrama(data)
        areas = gerar_areas(vor, regioes)
        
        df_areas = df_areas.append(areas, ignore_index=True)
        
    df_areas = df_areas[['Home_2', 'Home_3']]
    df_areas.dropna()
    df_areas = df_areas.loc[(df_areas['Home_2'] <= 1500) & (df_areas['Home_3'] <= 1500)]
            
    return df_areas
```
