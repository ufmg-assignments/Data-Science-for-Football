## Bibliotecas

```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean
import planarity
```

## Definindo as posições e criando o grafo

```python
# Define as coordenadas x e y de cada posição em campo
positions = {
    'Goleiro': (5.09,34.21),
    'Zagueiro1': (22.3025,44.23375),
    'Zagueiro2': (22.20125,27.5275),
    'Zagueiro3': (31.51625,56.3837500000001),
    'Zagueiro4': (29.08625,14.9725),
    'MeioCampo1': (40.31125,31.88125),
    'MeioCampo2': (47.715,47.7775000000001),
    'MeioCampo3': (61.52375,51.9912500000001),
    'MeioCampo4': (64.32125,29.04625),
    'MeioCampo5': (54.29625,12.64375),
    'Atacante': (78.9012500000001,33.50125)
}
```

```python
# Cria o grafo em NetworkX
G = nx.Graph()

# Adiciona os nós representando as posições em campo
for pos, coord in positions.items():
    G.add_node(pos, pos=coord)

# Cria as conexões entre as posições e adiciona as distâncias como peso das arestas
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 != node2 and not G.has_edge(node1, node2):
            coord1 = G.nodes[node1]['pos']
            coord2 = G.nodes[node2]['pos']
            dist = euclidean(coord1, coord2)
            G.add_edge(node1, node2, weight=dist)

# Define o layout do grafo
pos = nx.spring_layout(G, pos=positions)
plt.figure(figsize=(20, 14))
plt.xlim(0, 105)
plt.ylim(0, 70)

# Plota o grafo
pos = {node: data['pos'] for node, data in G.nodes(data=True)}
nx.draw(G, pos=pos)
plt.show()
```

## Árvore geradora mínima

```python
# Calcula a AGM usando o algoritmo de Kruskal
T = nx.minimum_spanning_tree(G)

pos = nx.spring_layout(T, pos=positions)
plt.figure(figsize=(20, 14))
plt.xlim(0, 105)
plt.ylim(0, 70)

# Plota o grafo
pos = {node: data['pos'] for node, data in T.nodes(data=True)}
nx.draw(T, pos=pos)
plt.show()
```

## Métricas de avaliação

```python
import pandas as pd

# Cálculo das medidas de centralidade
degree_centrality = nx.degree_centrality(C)
betweenness_centrality = nx.betweenness_centrality(C)
closeness_centrality = nx.closeness_centrality(C)
load_centrality = nx.load_centrality(C)

# Conversão do dicionário em tabela pandas
df = pd.DataFrame({
    'Grau': degree_centrality,
    'Intermediação': betweenness_centrality,
    'Proximidade': closeness_centrality,
    'Carga': load_centrality,
})

print(df)
```

```python
avg_clustering = nx.average_clustering(G)
print("Coeficiente de clusterização médio:", avg_clustering)
avg_clustering = nx.average_clustering(C)
print("Coeficiente de clusterização médio:", avg_clustering)
avg_clustering = nx.average_clustering(T)
print("Coeficiente de clusterização médio:", avg_clustering)
```

```python
cohesion = nx.algorithms.clique.graph_clique_number(G)
print("Coesão do grafo: ", cohesion)
cohesion = nx.algorithms.clique.graph_clique_number(C)
print("Coesão do grafo: ", cohesion)
cohesion = nx.algorithms.clique.graph_clique_number(T)
print("Coesão do grafo: ", cohesion)
```

## Detectando comunidades

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community

# Executa o algoritmo Louvain para detectar as comunidades do grafo
partition = community.greedy_modularity_communities(C)

# Plota o grafo, colorindo os vértices de acordo com a comunidade
# Plota o grafo
pos = {node: data['pos'] for node, data in C.nodes(data=True)}

colors = ['red', 'blue', 'green', 'orange']
for i, com in enumerate(partition):
    nx.draw_networkx_nodes(C, pos, nodelist=list(com), node_color=colors[i])
nx.draw_networkx_edges(C, pos)
plt.show()
```
